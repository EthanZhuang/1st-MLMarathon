## Why dataframe is so important?

1. 分析過程中產生的數據/結果儲存為結構化資料
    - ex1: 每筆交易資料彙總計算平均值、標準差等統計數值
2. 測試程式碼
    - 原始資料過於龐大，在相同結構的資料上**測試程式碼**是否能得到理想結果
    - 不確定**資料結構**的structure, 利用新建立的dataframe結構來去了解，而不是在原始資料上操作
